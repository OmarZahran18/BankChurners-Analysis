```python
#Number of Customers by Education Level
education_level = bank_churners['Education_Level'].value_counts()

colors = plt.cm.coolwarm(np.linspace(0, 1, len(education_level)))
bars = plt.bar(education_level.index, education_level.values, color=colors)

plt.title('Number of Customers by Education Level')
plt.xlabel('Education Level')
plt.ylabel("Numver Of Customers")
plt.xticks(rotation = 45)

for i, value in enumerate(education_level.values):
    plt.text(i, value + 50, str(value), ha='center')
    
plt.show()
```

##### Most customers are highly educated: The largest group of customers are graduates, followed by those with a high school education.

##### Customer numbers decrease with higher education: Generally, as the education level increases (beyond a graduate degree), the number of customers declines. The smallest group of customers holds a doctorate.

```python
#Customer Attrition Distribution'
attrition = bank_churners['Attrition_Flag'].value_counts()
plt.pie(attrition, labels=attrition.index, autopct='%1.1f%%', colors=['darkgreen','red'])
plt.title('Customer Attrition Distribution')
plt.show()

```

##### The vast majority of the customer base, 83.9%, are existing customers. This means that only a small portion, 16.1%, of customers have left the service.

```python
#Attrition by Gender
ax = sns.countplot(x='Gender', hue='Attrition_Flag', data=bank_churners, palette=['darkgreen','red'])
plt.title('Attrition by Gender')

for i in ax.patches:
    height = i.get_height() 
    ax.text(x=i.get_x() + i.get_width()/2, y=height + 5, s=int(height), ha='center')
    
plt.show()

```

##### Female customers have a slightly higher number of churned accounts than male customers, but gender does not appear to be a major factor in customer attrition since both genders have a similar proportion of existing and attrited customers.

```python
#Correlation between Numeric Columns
numeric_cols1 = ['Customer_Age', 'Credit_Limit', 'Total_Trans_Amt', 'Total_Trans_Ct' , 'Avg_Open_To_Buy' , 'Inactive_Months']

plt.figure(figsize=(8,6))
sns.heatmap(bank_churners[numeric_cols1].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Numeric Columns')
plt.show()
```

##### The chart shows that Total Transaction Amount and Total Transaction Count are strongly and positively correlated, meaning that as the number of transactions increases, the total value of those transactions also tends to increase. All other correlations between the variables (Customer Age, Credit Limit) are very weak.

```python
# Inactive Months, Avg Open To Buy, and Avg Utilization Ratio by Attrition Flag
numeric = ['Inactive_Months', 'Avg_Open_To_Buy', 'Avg_Utilization_Ratio']

fig, axes = plt.subplots(1, 3, figsize=(17,6))

for i, col in enumerate(numeric):
    sns.boxplot(x='Attrition_Flag', y=col, data=bank_churners, hue='Attrition_Flag', palette=['darkgreen','red'], ax=axes[i], legend=False)
    axes[i].set_title(f'{col} vs Attrition')
    axes[i].set_xlabel('Attrition_Flag')
    axes[i].set_ylabel(col)
plt.show()
```

##### Inactive Months: Customers who leave tend to have a higher average number of inactive months than existing customers. This suggests inactivity could be a sign of future attrition.
##### Credit Utilization: Attrited customers have a lower average credit utilization ratio, meaning they use a smaller portion of their available credit limit. This indicates that low engagement with the credit card is linked to higher churn.
##### Open to Buy: Existing customers have a higher average "open to buy" amount (available credit) compared to those who have left. This may be because existing customers tend to have higher credit limits.

```python
#Customer Age, Transaction Count, and Spending by Attrition Status

cols = ['Customer_Age', 'Total_Trans_Ct', 'Total_Trans_Amt']
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, col in enumerate(cols):
    sns.boxplot(
        x='Attrition_Flag',
        y=col,
        data=bank_churners,
        hue='Attrition_Flag',
        palette=['darkgreen', 'red'],
        ax=axes[i],
        legend=False
    )
    axes[i].set_title(f'{col} vs Attrition_Flag')
    axes[i].set_xlabel('')
    axes[i].set_ylabel(col)

plt.tight_layout()
plt.show()
```

 ##### The analysis showed that customers who leave the company tend to have significantly lower transaction counts and spending. Therefore, it is important to focus on engaging them with offers or monitoring low-usage customers before they churn

##### Recommendations
##### - The bank can focus on engaging customers with low activity (high Inactive_Months) to retain them.
##### - Monitoring customers with low card utilization (low Avg_Utilization_Ratio) may help reduce attrition.
##### - Analyzing the financial behavior of customers with low Total_Trans_Amt can assist in offering personalized promotions to retain them.
